```python
# Gráfico sobre os tempos de execução
blocosTempos1 = np.array([8645, 18225, 38122])
blocosTempos2 = np.array([7670, 10675, 21300])
blocosTempos4 = np.array([4629, 6196, 14908])
blocosTempos8 = np.array([3177, 4730, 10078])

# Definindo a largura das barras
barWidth = 0.2

# Aumentando o gráfico
plt.figure(figsize = (10,5))

# Definindo a posição das barras
g1 = np.arange(len(blocosTempos1))
g2 = [x + barWidth for x in g1]
g3 = [x + barWidth for x in g2]
g4 = [x + barWidth for x in g3]

# Criando as barras
plt.bar(g1, blocosTempos1, color='red', width=barWidth, label='1 Thread')
plt.bar(g2, blocosTempos2, color='green', width=barWidth, label='2 Threads')
plt.bar(g3, blocosTempos4, color='blue', width=barWidth, label='4 Threads')
plt.bar(g4, blocosTempos8, color='#6A5ACD', width=barWidth, label='8 Threads')

# Adicionando legendas as barras
plt.xlabel('Número de imagens ')
plt.ylabel('Tempo de execução (em ms)')
plt.xticks([r + barWidth for r in range(len(blocosTempos1))], ['25 imagens', '50 imagens', '100 imagens'])

# Criando titulo
plt.title('Algoritmo Concorrente em Blocos')

# Criando a legenda e exibindo o gráfico
plt.legend()
plt.show()
```
